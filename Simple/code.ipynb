```python
import pandas as pd
# pd.__version__
```

```python
#Read the CSV File into df
df = pd.read_csv('../clean.csv')
# df.head()
df
```

```python
#Calculate the number of views garnered by the 70th percentile job
m = df['views'].quantile(0.70)
m
#We can see that only 30% of the movies have gained more than 1788 views. Therefore, our value of m is 1788
```

```python
data = df
```

```python
# Calculate C
C = df['views'].mean()
C
```

```python
# Function to compute the Job weighted views for each position
def weighted_rating(x, m=m, C=C):
    v = x['views']
    R = x['saved']
    # Compute the weighted score
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Compute the score using the weighted_rating function defined above
data['score'] = data.apply(weighted_rating, axis=1)
```

```python
#Sort jobs in descending order of their views
data = data.sort_values('views', ascending=False)

#Print the top 20 jobs
data.tail(20)
```
