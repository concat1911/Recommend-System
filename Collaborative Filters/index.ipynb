```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows=500
pd.options.display.max_columns=500
```

```python
movies=pd.read_csv("./u.item",sep="|",encoding="latin-1",header=None)
movies=movies.iloc[:,0:2]
movies.columns=["Movie_Id","Movie_Name"]
movies.head(20)
```

```python
ratings=pd.read_csv("./u.data",sep="\t",header=None) 
ratings.columns=["User_Id","Movie_Id","Rating","Timestamp"]
ratings = ratings.drop("Timestamp", axis=1)
ratings.head(20)
```

```python
# Merge ratings and movies files with movie_id key for names of movies
ratings=pd.merge(ratings,movies,on="Movie_Id",how="inner")
ratings.head()
```

```python
# Creating user-item matrix because we're building collaborative filtering recommendations
user_item_matrix=ratings.pivot_table(index=["Movie_Id"],columns=["User_Id"],values="Rating").reset_index(drop=True)
user_item_matrix.fillna(0,inplace=True)
user_item_matrix.head()
```

```python
# Similarities Cal.
from sklearn.metrics.pairwise import pairwise_distances
from scipy.spatial.distance   import correlation,cosine
# https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.spatial.distance.correlation.html
```

```python
similarity_matrix = (1 - pairwise_distances(user_item_matrix, metric="correlation"))
```

```python
movie_name=movies["Movie_Name"][207]
movie_id=movies[movies["Movie_Name"]==movie_name].index
movie_id=movie_id[0]
movie_name
```

```python
movies["Similarity"]=similarity_matrix[movie_id]
```
